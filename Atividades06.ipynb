```python
from scipy.integrate import simpson
from numpy import trapz
from scipy.interpolate import CubicSpline

import sympy as smp
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Atividade 1 - Para a função y = 2.0*sin^2(3.0*x + 1.77)*cos(1.0*x), no intervalo de [0,2PI] faça:
# (a) gere um conjunto de dados incluindo ruído;
# (b) Faça uma suavização ou interpolação (com resultados em gráfico);
# (c) Calcule o valor da integral pelos métodos do trapézio e Simpson;
# (d) Compare com o valor exato da integral;
```

Atividade 01:

Letra a)

```python
def funcao(x):
    return 2 * ((np.sin(3*x + 1.77))**2) * np.cos(x)
    
X = np.linspace(0, 2*np.pi, 25)
Y = funcao(X)

ruido = 0.13*np.random.normal(size=Y.size)
Y = Y + ruido

plt.scatter(X, Y, marker='+', color='green', label='Dados')
plt.legend(loc='best')
plt.title('Dados');
```

Letra B) Suavização Cubic Spline

```python
x_cs = np.linspace(0, 2*np.pi, 1300)
y_cs = funcao(x_cs)
y_smooth = CubicSpline(x_cs, y_cs)

plt.scatter(X, Y, marker='+', color='green', label = 'Dados')
plt.plot(x_cs, y_smooth(x_cs), linestyle='-', color='#808000', label='Suavização Cubic Spline')
plt.legend(loc='best')
plt.title('Dados Suavizados');
```

Letra C) Método do Trapézio


```python
x = np.arange(0, 2*np.pi, 0.01)
y = funcao(x)

area_tp = trapz(y, x)
print ('Área obtida pela integração por trapézio = ',area_tp)
```

Letra C) Método Simpsons


```python
area_s = simpson(y=y, x=x)
print ('Área obtida pela integração por Simpson = ', area_s)
```

Letra D) Integral real


```python
integral = 2*((3*np.sin(177 / 50))/35 - (5*np.sin((700*np.pi + 177)/50) + 7*np.sin((500*np.pi + 177)/50)) / 140)
print ('Área obtida pela integração real = ', integral)
```

```python
# Atividade 2 - Para a função y = 1/[x^2 * sqrt(4 - x^2)];, no intervalo de [0,20] faça:
# (a) gere um conjunto de dados incluindo ruído;
# (b) Faça uma suavização ou interpolação (com resultados em gráfico);
# (c) Calcule o valor da integral pelos métodos do trapézio e Simpson;
# (d) Compare com o valor exato da integral;
```

Atividade 2:

Letra A) Dados

```python
def raiz(x):
    return 1/(x**2 * np.sqrt(4 - x**2))

X = np.linspace(-1.9, 1.9, 20)
Y = raiz(X)
ruido = 0.13*np.random.normal(size=Y.size)
Y = Y + ruido

plt.scatter(X, Y, marker='+', color='#836953', label='Dados')
plt.ylim(0, 3)
plt.legend(loc='best')
plt.title('Dados');
```

Letra B) Interpolando com cubic spline

```python
x_cs = np.linspace(-1.9, 1.9, 1000)
y_cs = raiz(x_cs)
y_smooth = CubicSpline(x_cs, y_cs)

plt.scatter(X, Y, marker='+', color='#836953' , label='Dados')
plt.plot(x_cs, y_smooth(x_cs), linestyle='-', color='#e9967a', label='Suavização Cubic Spline')

plt.ylim(0, 3)
plt.legend(loc='best')
plt.title('Dados Suavizados');
```

Letra c) Método Trapézio

```python
x = np.arange(-1.9, 1.9, 0.01)
y = raiz(x)

area_tp = trapz(y, x)
print ('Área obtida pela integração por trapézio = ',area_tp)
```

Letra C) Método Simpson


```python
area_s = simpson(y=y, x=x)
print ('Área obtida pela integração por Simpson = ', area_s)
```

Letra D) Integral real


```python
integral_raiz = -(np.sqrt(4-(1.9)**2)/4*(1.9)) + (np.sqrt(4-(-1.9)**2)/4*(-1.9))
print ('Área obtida pela integração real = ', integral_raiz)
```
