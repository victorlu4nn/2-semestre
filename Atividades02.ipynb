```python
# Atividade 1 - Ajuste de uma curva do tipo : y = a * sin(b*x + c)
# Usando curve_fit from Scipy, modifique os exemplos dados na aula anterior 
# para fazer o ajuste pedido
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import sympy 
from scipy.optimize import curve_fit
```

```python
def sen(x, a, b, c):
    return a * np.sin(b * x + c)

X = np.linspace(-10, 10, 1000)
Y = sen(X, 1.3, 2, 1.5)

# Adicionar ruído
ruido = 0.13 * np.random.normal(size=Y.size)
Y = Y + ruido

# Gráfico dos dados com o ruído
fig, ax = plt.subplots()
ax.scatter(X, Y, s=20, color='#006400', marker='+')
ax.set_facecolor('#f5f5f5')

# Ajuste de dados
# Corrigido para incluir três valores em `p0` e limites consistentes em `bounds`
pars, cov = curve_fit(f=sen, xdata=X, ydata=Y, p0=[1.2, 2, 1.0], bounds=([0, 0, 0], [2, 3, np.pi]))

# Plot do ajuste
ax.plot(X, sen(X, *pars), linestyle='--', linewidth=1.3, color='black', label="Ajuste")
ax.legend()

plt.show()
```

```python
# Atividade 2 - Ajuste de uma curva do tipo : y = a/( (x - b)**2 + c )
# Usando curve_fit from Scipy, modifique os exemplos dados na aula anterior para fazer o ajuste pedido
```

```python
def funcao(x, a, b, c):
    return a/((x-b)**2+c)

X = np.linspace(-10,10,100)
Y = funcao(X, 20, 0.13, 1.3)

ruido = 0.8 * np.random.normal(size=Y.size)
Y = Y + ruido

fig, ax = plt.subplots()
ax.scatter(X, Y, s=15, color='#722f37', marker=".")
ax.set_facecolor('#f5f5f5')

pars, cov = curve_fit(f=funcao, xdata=X, ydata=Y, p0=[20, 0.1, 1.8])
ax.plot(X, funcao(X, *pars), linestyle='-.', linewidth=2,color='black')

plt.show()
```

```python
# Atividade 3 - Ajuste de uma curva do tipo : y = a* sin(b*x)*sqrt(x+c)
# Usando curve_fit from Scipy, modifique os exemplos dados na aula anterior 
# para fazer o ajuste pedido
```

```python
def curva(x, a, b, c):
    return a * np.sin(b * x) * np.sqrt(x + c)


X = np.linspace(0, 10, 1000)
Y = curva(X, 2.1, 1.2, 2)

ruido = 0.9 * np.random.normal(size=Y.size)
Y = Y + ruido

fig, ax = plt.subplots()
ax.scatter(X, Y, s=5, color="red", marker="d")
ax.set_facecolor("#f5f5f5")

pars, cov = curve_fit(f=curva, xdata=X, ydata=Y, p0=[2.0, 1.1, 1.8], bounds=([0, 0, 0], [np.inf, np.inf, np.inf]))

ax.plot(X, curva(X, *pars), linestyle='-.', linewidth=1.52, color='black', label="Ajuste")
ax.legend()

plt.show()
```
