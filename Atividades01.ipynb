```python
# 1) Faça um gráfico da função y = (1/x^2) * sen(3*x), assim como de suas derivadas de ordem um e dois. (2.0 pontos)
####################################################################################################################
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(0.1,0.3,100)

y1 =(1/(x**2)) *10* np.sin(30*x)
y2 = -(2*np.sin(3*x)-3*x*np.cos(3*x))/(x**3)
y3 = -((9*x**2-6)*np.sin(3*x)+12*x*np.cos(3*x))/(x**4)

fig, ax = plt.subplots()
ax.plot(x, y1, color="blue", label="y(x)")
ax.plot(x, y2, color="red", label="y'(x)")
ax.plot(x, y3, color="green", label="y”(x)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```

```python
# 2) Através de análise gráfica localize as raízes de x**3-x**2+1 = x+2.  
# Faça um gráfico mostrando, via um "inset", uma das raízes. (2.0 pontos)
```

```python
def f(x):
    return x**3 - x**2 + 1

def g(x):
    return x + 2

# Função para plotar e formatar eixos
def plot_and_format_axes(ax, x, fontsize):
    ax.plot(x, f(x), color="cyan", linewidth=2, label="f(x)")
    ax.plot(x, g(x), color="blue", linewidth=2, label="g(x)")
    ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5)) #define o número máximo de divisões do eixo x
    ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4)) #define o número máximo de divisões do eixo x
    ax.set_xlabel(r"$x$", fontsize=fontsize)
    ax.set_ylabel(r"$y$", fontsize=fontsize)

x = np.linspace(-1, 3, 100)

# Gráfico principal
fig, ax = plt.subplots()
plot_and_format_axes(ax, x, 18)

# Linhas verticais no gráfico principal
x0, x1 = 1.7, 2.0
ax.axvline(x0, ymax=0.4, color="grey", linestyle=":")
ax.axvline(x1, ymax=0.4, color="grey", linestyle=":")

# Gráfico de detalhe (zoom)
ax_insert = fig.add_axes([0.25, 0.55, 0.3, 0.3], facecolor="none")
x_detail = np.linspace(x0, x1, 100)
plot_and_format_axes(ax_insert, x_detail, 14)

# Exibe o gráfico
plt.show()

#Conclusão: a raiz deve estar entre 1,75 e 1,9
```

```python
# 3) Faça gráficos 3D de z = sen(x)^2 * (3*y - 3)**2 do tipo "surface", "wireplot" e "contour". 
# O gráfico deverá ser no intervalo
# -10 < x < 10 e -10 < y < 10. (2.0 pontos)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4), subplot_kw={'projection': '3d'})

def title_and_labels(ax, title):
    ax.set_title(title)
    ax.set_xlabel("$x$", fontsize=16)
    ax.set_ylabel("$y$", fontsize=16)
    ax.set_zlabel("$z$", fontsize=16)

x = y = np.linspace(-10, 10, 74)
X, Y = np.meshgrid(x, y)
Z = (np.sin(X))**2 * (3*Y - 3)**2
norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

# PRIMEIRO PLOT
p = axes[0].plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=0, antialiased=False, norm=norm, cmap=mpl.cm.viridis)
cb = fig.colorbar(p, ax=axes[0], shrink=0.6)
title_and_labels(axes[0], "plot_surface")

# SEGUNDO PLOT
p = axes[1].plot_wireframe(X, Y, Z, rstride=2, cstride=2, color="darkgrey")
title_and_labels(axes[1], "plot_wireframe")

# TERCEIRO PLOT
cset = axes[2].contour(X, Y, Z, zdir='z', offset=0, norm=norm, cmap=mpl.cm.Blues)
cset = axes[2].contour(X, Y, Z, zdir='y', offset=3, norm=norm, cmap=mpl.cm.Blues)
title_and_labels(axes[2], "contour")

plt.show()
```

```python
# 4) Faça um gráfico tipo contour plot da função de Rosenbrock: f(x,y) = (a - x)**2 + b(y - x**2)**2 com a = 1 e b = 100. 
# O intervalo será de -3 < x < 3  e -3 < y < 3. (2.0 pontos)
```

```python
from mpl_toolkits.mplot3d import Axes3D

a = 1
b = 100

fig, axes = plt.subplots(1, 1, figsize=(16, 4), subplot_kw={'projection': '3d'})

def title_and_labels(ax, title):
    ax.set_title(title)
    ax.set_xlabel("$x$", fontsize=16)
    ax.set_ylabel("$y$", fontsize=16)
    ax.set_zlabel("$z$", fontsize=16)

x = y = np.linspace(-3, 3, 74)
X, Y = np.meshgrid(x, y)
Z = (a - X)**2 + b*(Y - X**2)**2
norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

# Use 'axes' diretamente, pois não é uma lista
cset = axes.contour(X, Y, Z, zdir='z', offset=0, norm=norm, cmap=mpl.cm.Blues)
cset = axes.contour(X, Y, Z, zdir='y', offset=3, norm=norm, cmap=mpl.cm.Blues)
title_and_labels(axes, "contour")

plt.show()
```

```python
# 5) Faça um gráfico tipo "surface" da função de Rosenbrock: f(x,y) = (a - x)**2 + b(y - x**2)**2 com a = 1 e b = 100. 
# O intervalo será de -3 < x < 3  e -3 < y < 3. (2.0 pontos)
```

```python
a= 1
b = 100

fig, axes = plt.subplots(1, 1, figsize=(16, 4), subplot_kw={'projection': '3d'})

def title_and_labels(ax, title):
    ax.set_title(title)
    ax.set_xlabel("$x$", fontsize=16)
    ax.set_ylabel("$y$", fontsize=16)
    ax.set_zlabel("$z$", fontsize=16)

x = y = np.linspace(-3, 3, 74)
X, Y = np.meshgrid(x, y)
Z = (a - X)**2 + b*(Y - X**2)**2
norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

## Criação do gráfico de superfície
surface = axes.plot_surface(X, Y, Z, cmap=mpl.cm.Blues, norm=norm)

# Adiciona uma barra de cores para o gráfico de superfície
fig.colorbar(surface, ax=axes, shrink=0.5, aspect=5)
title_and_labels(axes, "Surface Plot")

plt.show()
```
